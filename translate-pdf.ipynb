```python
page_count = 1
for page in pages:
    document = doctran.parse(content=str(page))
    document = document.translate(language="english").execute()
    new_file_path = f"docs/translated-{page_count}.md"
    with open(new_file_path, 'w') as file:
        file.write(document.transformed_content)
    
    page_count += 1
    if page_count == 10:
        break
```
